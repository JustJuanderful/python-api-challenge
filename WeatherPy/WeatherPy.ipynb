```python
#import dependencies
import os
import csv
import sys
assert sys.version_info >= (3, 5)

import sklearn
assert sklearn.__version__ >= "0.20"
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

import requests

# Import API key
from config import api_key

from citipy import citipy

import numpy as np
import pandas as pd

# To plot figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```python
# Generate City List

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# Initiate Lists for cities and lat_long
cities = []
lat_lng = []

# Generate random lat and long combos
lat = np.random.uniform(low=-90.000, high=90.000, size=1500)
lng = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lng = list(zip(lat, lng))

# Find nearest city to random coordinates
for coord in lat_lng:
    city_name = citipy.nearest_city(coord[0], coord[1]).city_name
    
    if city_name not in cities:
        cities.append(city_name)

num_cities = len(cities)

if num_cities >= 500:
    print(f'Juanderful! You have at least 500 cities. As matter of fact, you have {num_cities}!')
else:
    print('Try again.')
```

```python
# Set up lists to hold reponse info
temperature = []
humidity = []
cloudiness = []
wind_speed = []
lat = []
lng = []
country = []
datetime = []
query_cities = []

# Initialize city counter
city_counter = 0

# Save config information.
base_url = "http://api.openweathermap.org/data/2.5/weather?q="

# Prompt beginning of data retrieval
print('-'*15)
print('City Data Retrieval In-Progress')
print('-'*15)

# Loop through the list of cities and perform a request for data on each
for city in cities:
    query_url = base_url + city + "&appid=" + api_key + '&units=imperial'
    response = requests.get(query_url).json()
    
    # Increment/record data retrieval for city
    city_counter += 1
    print(f'Retrieving city {city_counter} data... {city}')
    
    try:
        temperature.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        country.append(response['sys']['country'])
        datetime.append(response['dt'])
        query_cities.append(city)
    except:
        pass

# Prompt end of data retrieval
print('-'*15)
print('City Data Retrieval Complete')
print('-'*15)
```

```python
# Create DataFrame from API response
weatherpy_df = pd.DataFrame({"City": query_cities,
                        "Temperature (F)": temperature,
                        "Humidity (%)": humidity,
                        "Cloudiness (%)": cloudiness,
                        "Wind Speed (mph)": wind_speed,
                        "Latitude": lat,
                        "Longditude" : lng,
                        "Country" : country,
                        "Date_Time" : datetime
                       })
weatherpy_df.head(10)
```

```python
# Save output to CSV

# Output csv path
output_file = os.path.join('output_data', 'output_cities.csv')

# Open and write to output csv
with open(output_file, 'w') as datafile:
    weatherpy_df.to_csv(output_file)
```

```python
# Split weatherpy_df into nothern and southern hemisphere for Linear Regression

# Northern hemisphere df
northern_df = weatherpy_df[weatherpy_df['Latitude'] >= 0]
northern_df.reset_index(drop=True).head(10)
```

```python
# Southern hemisphere df
southern_df = weatherpy_df[weatherpy_df['Latitude'] < 0]
southern_df.reset_index(drop=True).head(10)
```

```python
# Linear Regression from sklearn
linr = LinearRegression()

# Hemispheres
northern = northern_df
southern = southern_df

# Define regression plot pair function
def regression_plots(hem1, hem2, pred_val, crit_val):

    X1 = hem1[pred_val].to_frame().reset_index(drop=True)
    y1 = hem1[crit_val].to_frame().reset_index(drop=True)
    linr.fit(X1, y1) # Regression for hemisphere 1 (Northern Hemisphere)
    
    X2 = hem2[pred_val].to_frame().reset_index(drop=True)
    y2 = hem2[crit_val].to_frame().reset_index(drop=True)
    linr.fit(X2, y2) # Regression for hemisphere 2 (Southern Hemisphere)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5)) # Create subplots
    
    # Plot Northern Hemisphere regression (hem1)
    ax1.scatter(X1, y1)
    ax1.plot(X1, linr.predict(X1), color='r')
    ax1.set_ylabel(crit_val)
    ax1.set_xlabel(pred_val)
    ax1.set_title('Northern Hemisphere', fontsize=14)
    
    # Plot Southern Hemisphere regression (hem2)
    ax2.scatter(X2, y2)
    ax2.plot(X2, linr.predict(X2), color='r')
    ax2.set_ylabel(crit_val)
    ax2.set_xlabel(pred_val)
    ax2.set_title('Southern Hemisphere', fontsize=14)
    
    # Northern Hemisphere Stats
    print('Northern Hemisphere Stats')
    print('-'*20)
    # Coefficients
    print('Coefficients: \n', linr.fit(X1, y1).coef_)
    # MSE
    print('Mean squared error: %.2f'
          %mean_squared_error(y1, linr.predict(X1)))
    # The coefficient of determination: 1 is perfect prediction
    print('Coefficient of determination: %.2f'
          % r2_score(y1, linr.predict(X1)))
    print('-'*20)
    
    # Southern Hemisphere Stats
    print('Southern Hemisphere Stats')
    print('-'*20)
    # Coefficients
    print('Coefficients: \n', linr.fit(X2, y2).coef_)
    # MSE
    print('Mean squared error: %.2f'
        % mean_squared_error(y2, linr.predict(X2)))
    # The coefficient of determination: 1 is perfect prediction
    print('Coefficient of determination: %.2f'
          % r2_score(y2, linr.predict(X2)))
    print('-'*20)
```

```python
# Temperature (F) vs. Latitude Regression
regression_plots(northern, southern, 'Latitude', 'Temperature (F)')

# Save regression plots image
plt.savefig('output_data/images/Temp_v_Lat_r.png')

# Note - not sure why the line of fit on the first chart below is acting like that.
```
